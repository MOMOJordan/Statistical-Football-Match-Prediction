#### remplacer les nan pour les vars de gk

```python
gk_mt_columns = all_columns_of_interest_gk_common_mt[:]
nan_before = data_home_player[gk_mt_columns].isnull().sum(axis=1)
#plt.hist(nan_before)
print(nan_before.sum())
```

```python
df_filled_gk_mt = data_home_player[gk_mt_columns+['ID', 'POSITION']]

# compute means per ID for only goalkeeper
gk_data = data_home_player[gk_mt_columns+['ID', 'POSITION']][data_home_player['POSITION'] == 'goalkeeper']
mean_values = gk_data.groupby('ID')[gk_mt_columns].mean()

# fill goalkeeper nan by the means
df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] == 'goalkeeper', gk_mt_columns] = df_filled_gk_mt[df_filled_gk_mt['POSITION'] == 'goalkeeper'].groupby('ID')[gk_mt_columns].transform(lambda x: x.fillna(x.mean()))

# fill other nan by zero
df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] != 'goalkeeper', gk_mt_columns] = df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] != 'goalkeeper', gk_mt_columns].fillna(0)
```

```python
nan_after = df_filled_gk_mt[gk_mt_columns][df_filled_gk_mt.POSITION=='goalkeeper'].isnull().sum(axis=1)
#plt.hist(nan_after)
print(nan_after.sum())
```

```python
all_gk_means = df_filled_gk_mt[df_filled_gk_mt.POSITION == 'goalkeeper'].mean(axis=0, numeric_only=True)
df_filled_gk_mt.fillna(all_gk_means, inplace=True)
```

```python
nan_after = df_filled_gk_mt[gk_mt_columns][df_filled_gk_mt.POSITION=='goalkeeper'].isnull().sum(axis=1)
print(nan_after.sum())
```
