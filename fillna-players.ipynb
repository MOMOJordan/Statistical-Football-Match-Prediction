### Filling missing values

```python
base_columns_of_interest_players_common_mt = [
'PLAYER_ACCURATE_CROSSES',
'PLAYER_ACCURATE_PASSES',
'PLAYER_AERIALS_WON',
'PLAYER_ASSISTS',
'PLAYER_BIG_CHANCES_CREATED',
'PLAYER_BIG_CHANCES_MISSED',
'PLAYER_BLOCKED_SHOTS',
'PLAYER_CAPTAIN',
'PLAYER_CLEARANCES',
'PLAYER_CLEARANCE_OFFLINE',
'PLAYER_DISPOSSESSED',
'PLAYER_DRIBBLED_ATTEMPTS',
'PLAYER_DRIBBLED_PAST',
'PLAYER_DUELS_LOST',
'PLAYER_DUELS_WON',
'PLAYER_ERROR_LEAD_TO_GOAL',
'PLAYER_FOULS',
'PLAYER_FOULS_DRAWN',
'PLAYER_GOALS',
'PLAYER_GOALS_CONCEDED',
'PLAYER_HIT_WOODWORK',
'PLAYER_INTERCEPTIONS',
'PLAYER_KEY_PASSES',
'PLAYER_MINUTES_PLAYED',
'PLAYER_OFFSIDES',
'PLAYER_OWN_GOALS',
'PLAYER_PASSES',
'PLAYER_PENALTIES_COMMITTED',
'PLAYER_PENALTIES_MISSES',
'PLAYER_PENALTIES_SAVED',
'PLAYER_PENALTIES_SCORED',
'PLAYER_PENALTIES_WON',
'PLAYER_REDCARDS',
'PLAYER_SHOTS_BLOCKED',
'PLAYER_SHOTS_ON_TARGET',
'PLAYER_SHOTS_TOTAL',
'PLAYER_STARTING_LINEUP',
'PLAYER_SUCCESSFUL_DRIBBLES',
'PLAYER_TACKLES',
'PLAYER_TOTAL_CROSSES',
'PLAYER_TOTAL_DUELS',
'PLAYER_YELLOWCARDS',
'PLAYER_PUNCHES',
'PLAYER_LONG_BALLS',
'PLAYER_LONG_BALLS_WON',
'PLAYER_SHOTS_OFF_TARGET'
]

base_columns_of_interest_gk_common_mt = [
'PLAYER_GOALKEEPER_GOALS_CONCEDED',
'PLAYER_SAVES',
'PLAYER_SAVES_INSIDE_BOX'
]

base_columns_of_interest_players_only_avg_std_mt = [
'PLAYER_ACCURATE_PASSES_PERCENTAGE',
'PLAYER_RATING'
]

all_format_vars = ['_season_sum', '_season_average', '_season_std', '_5_last_match_sum', '_5_last_match_average', '_5_last_match_std']
avg_std_format_vars = ['_season_average', '_season_std', '_5_last_match_average', '_5_last_match_std']
```

```python
all_columns_of_interest_players_common_mt = []
all_columns_of_interest_gk_common_mt = []

for format_ in all_format_vars:
    all_columns_of_interest_players_common_mt+= [i+format_ for i in base_columns_of_interest_players_common_mt]
    all_columns_of_interest_gk_common_mt+= [i+format_ for i in base_columns_of_interest_gk_common_mt]

all_columns_of_interest_players_only_avg_std_mt = []
for format_ in avg_std_format_vars:
    all_columns_of_interest_players_only_avg_std_mt+= [i+format_ for i in base_columns_of_interest_players_only_avg_std_mt]
```

```python
players_mt_columns_no_full_nan = [column for column in players_mt_columns if column not in columns_full_nan] 
```

```python
df_filled_players_mt = data_home_player[players_mt_columns_no_full_nan+['ID']].fillna(data_home_player[players_mt_columns_no_full_nan+['ID']].groupby('ID').transform('mean', numeric_only = True))
```

```python
nan_after = df_filled_players_mt[players_mt_columns_no_full_nan].isnull().sum(axis=1)
#plt.hist(nan_after)
print(nan_after.sum())
```

S'il reste encore des valeurs manquantes à cette étape, cela veut dire qu'il y a des équipes entières pour lesquels toute une donnée est manquante. Dans ce cas on va fillna le reste par la moyenne de tous les joueurs du dataset, cela fait un peu sens dans la mesure où la moyenne de toutes les équipes du monde (ici le dataset) peut représenter la performance lambda d'un joueur de foot à cette stats.

```python
df_filled_players_mt[players_mt_columns_no_full_nan] = df_filled_players_mt[players_mt_columns_no_full_nan].fillna(df_filled_players_mt[players_mt_columns_no_full_nan].mean())
```

```python
nan_after = df_filled_players_mt[players_mt_columns_no_full_nan].isnull().sum(axis=1)
print(nan_after.sum())
```

#### remplacer les nan pour les vars de gk

```python
gk_mt_columns = all_columns_of_interest_gk_common_mt[:]
nan_before = data_home_player[gk_mt_columns].isnull().sum(axis=1)
#plt.hist(nan_before)
print(nan_before.sum())
```

```python
df_filled_gk_mt = data_home_player[gk_mt_columns+['ID', 'POSITION']]

# compute means per ID for only goalkeeper
gk_data = data_home_player[gk_mt_columns+['ID', 'POSITION']][data_home_player['POSITION'] == 'goalkeeper']
mean_values = gk_data.groupby('ID')[gk_mt_columns].mean()

# fill goalkeeper nan by the means
df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] == 'goalkeeper', gk_mt_columns] = df_filled_gk_mt[df_filled_gk_mt['POSITION'] == 'goalkeeper'].groupby('ID')[gk_mt_columns].transform(lambda x: x.fillna(x.mean()))

# fill other nan by zero
df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] != 'goalkeeper', gk_mt_columns] = df_filled_gk_mt.loc[df_filled_gk_mt['POSITION'] != 'goalkeeper', gk_mt_columns].fillna(0)
```

```python
nan_after = df_filled_gk_mt[gk_mt_columns][df_filled_gk_mt.POSITION=='goalkeeper'].isnull().sum(axis=1)
#plt.hist(nan_after)
print(nan_after.sum())
```

```python
all_gk_means = df_filled_gk_mt[df_filled_gk_mt.POSITION == 'goalkeeper'].mean(axis=0, numeric_only=True)
df_filled_gk_mt.fillna(all_gk_means, inplace=True)
```

```python
nan_after = df_filled_gk_mt[gk_mt_columns][df_filled_gk_mt.POSITION=='goalkeeper'].isnull().sum(axis=1)
print(nan_after.sum())
```
